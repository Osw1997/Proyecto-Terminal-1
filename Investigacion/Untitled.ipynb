```python
import pandas as pd
import requests
import time
import numpy as np
```

```python
df = pd.read_csv('LODAtlas_geourls_wSparqlEndpoint - copia.csv', encoding='latin-1', sep='@')
```

```python
payload = {}
headers = {
  'Connection': 'keep-alive',
  'Accept': 'application/json, text/javascript, */*; q=0.01',
  'X-Requested-With': 'XMLHttpRequest',
  'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.122 Safari/537.36',
  'Content-Type': 'application/json; charset=utf-8',
  'Referer': 'http://lodatlas.lri.fr/',
  'Accept-Language': 'es-US,es;q=0.9,en-US;q=0.8,en;q=0.7,es-419;q=0.6'
}

list_se = []

for repo,name in zip(df['Nombre repositorio'], df['Nombre dataset']):
    try:
        url = "http://lodatlas.lri.fr/lodatlas/rest/repos/" + repo + "/datasets/" + name
        request_SE = requests.get(url, headers=headers, data=payload).json()
#         print(request_SE)
        resources = request_SE['ckan']['resources']
        # Se recorre cada recurso con el fin de encontrar el SPARQL endpoint
        flag = False
        for resource in resources:
#             if resource['format'].upper() == 'SPARQL' or resource['format'].lower == 'api/sparql':
            if resource['verifiedFormat'].upper() == 'SPARQL':
                print('El SPARQL endpoint para repo:' + repo + ' dataset: ' + name + ' es: ' + resource['url'])
                flag = True
                list_se.append(resource['url'])
                break
        if not flag:
            print('No existe SPARQL endpoint para repo:' + repo + ' dataset: ' + name)
            list_se.append('-')
        time.sleep(2)
        print('=' * 20)
    except Exception as e:
        print('Error Sparql endpoint')
        print('URL: ' + url)
        print(e)
#         time.sleep(2)
```
